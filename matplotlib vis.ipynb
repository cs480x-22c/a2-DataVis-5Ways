```python
import pandas as pd
df = pd.read_csv("cars-sample.csv")

def mapColor(m):
    if m == "ford":
        return (.75, .75, .45, .75) 
    elif m == "toyota":
        return (.95, .65, .95, .75)
    elif m == "bmw":
        return (.95, .70, .70, .75) 
    elif m == "honda":
        return (.55, .85, .75, .75) 
    elif m == "mercedes":
        return (.60, .80, .95, .75)
    
color = df.Manufacturer.apply(lambda x: mapColor(x))
size = df.Weight.apply(lambda x: x / 25)

import matplotlib.pyplot as plt
plt.scatter(df.Weight, df.MPG, color=color, s=size)
plt.xlabel("Weight")
plt.xlim(1600, 5100)
plt.xticks(ticks=[2000, 3000, 4000, 5000])
plt.ylabel("MPG")
plt.ylim(9, 46)
plt.yticks(ticks=[10, 20, 30, 40])
plt.show()
```
