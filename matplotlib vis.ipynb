```python
import pandas as pd
df = pd.read_csv("cars-sample.csv")

def mapColor(m):
    if m == "ford":
        return (.75, .75, .45, .6) 
    elif m == "toyota":
        return (.95, .65, .95, .6)
    elif m == "bmw":
        return (.95, .70, .70, .6) 
    elif m == "honda":
        return (.55, .85, .75, .6) 
    elif m == "mercedes":
        return (.60, .80, .95, .6)
    
color = df.Manufacturer.apply(lambda x: mapColor(x))
size = df.Weight.apply(lambda x: x / 25)

import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot()
ax.scatter(df.Weight, df.MPG, color=color, s=size)
ax.set_xlabel("Weight")
ax.set_xlim(1600, 5100)
ax.set_xticks(ticks=[2000, 3000, 4000, 5000])
ax.set_ylabel("MPG")
ax.set_ylim(9, 46)
ax.set_facecolor("#ECECEC")
ax.set_yticks(ticks=[10, 20, 30, 40]);
```
