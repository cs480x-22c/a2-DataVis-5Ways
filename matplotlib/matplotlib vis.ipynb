```python
import pandas as pd
df = pd.read_csv("../cars-sample.csv")

def mapColor(m):
    if m == "ford":
        return (180/255, 180/255, 75/255, .5) 
    elif m == "toyota":
        return (240/255, 100/255, 240/255, .5)
    elif m == "bmw":
        return (240/255, 135/255, 135/255, .5) 
    elif m == "honda":
        return (40/255, 210/255, 175/255, .5) 
    elif m == "mercedes":
        return (80/255, 200/255, 245/255, .5)
    
color = df.Manufacturer.apply(lambda x: mapColor(x))
size = df.Weight.apply(lambda x: 3.14 * (x / 500)**2)

import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot()
ax.grid(color="white")
ax.scatter(df.Weight, df.MPG, color=color, s=size)
ax.set_xlabel("Weight")
ax.set_xlim(1600, 5100)
ax.set_xticks(ticks=[2000, 3000, 4000, 5000])
ax.set_ylabel("MPG")
ax.set_ylim(9, 46)
ax.set_facecolor("#ECECEC")
ax.set_yticks(ticks=[10, 20, 30, 40])

line1 = Line2D(range(1), range(1), color="white", marker='o', markerfacecolor=(240/255, 135/255, 135/255))
line2 = Line2D(range(1), range(1), color="white", marker='o', markerfacecolor=(180/255, 180/255, 75/255))
line3 = Line2D(range(1), range(1), color="white", marker='o', markerfacecolor=(40/255, 210/255, 175/255))
line4 = Line2D(range(1), range(1), color="white", marker='o', markerfacecolor=(80/255, 200/255, 245/255))
line5 = Line2D(range(1), range(1), color="white", marker='o', markerfacecolor=(240/255, 100/255, 240/255))
ax.legend((line1,line2,line3,line4,line5),('bmw','ford', 'honda', 'mercedes', 'toyota'),numpoints=1, loc=1,title="Manufacturer")
```
