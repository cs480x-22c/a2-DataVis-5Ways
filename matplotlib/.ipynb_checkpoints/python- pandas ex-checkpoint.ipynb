```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib import colors

#Read Data
df = pd.read_csv('cars-sample.csv', index_col =0)

#Normalize for scaling
copydf = df.copy()
copydf['Weight']= df['Weight'] / 100
print(copydf['Weight']);

#Color mapping
colormap =  matplotlib.colors.ListedColormap(matplotlib.cm.get_cmap("Set1").colors[:5])
copydf['Manufacturer'] = copydf.Manufacturer.astype('category').cat.codes

#Plot
plt.figure(figsize=(10,8))
plt.scatter(x=df.Weight, y=df.MPG, s=copydf.Weight, c=copydf.Manufacturer.astype('category').cat.codes, cmap=colormap, alpha=.5)
plt.xlabel("Weight")
plt.ylabel("MPG")


#Legend
#plt.legend(handles=scatter.legend_elements()[0], labels=copydf.Weight, title="Weight")
cbar = plt.colorbar(ticks= [.4, 1.2, 2.0, 2.8, 3.6])
cbar.ax.set_yticklabels(['bmw', 'ford', 'honda', 'mercedes', 'toyota']) 
plt.show()
```
