```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
import seaborn as sns

cars = pd.read_csv("cars-sample.csv")
```

```python
colors = dict(bmw="#ed1c16", ford="#ffff66", honda="#31cb1b", mercedes="#17d4fc", toyota="#ED8AF8")

ax = sns.scatterplot(data=cars, x="Weight", y="MPG", linewidth=0,
                     size="Weight",  sizes=(70,300), hue="Manufacturer", palette=colors, alpha=0.5, legend='brief')

plt.title("Data on Various Cars")
plt.xlim([1500, 5050])
plt.ylim([8, 47])
plt.xticks(np.arange(2000, 6000, 1000), fontweight='bold')
plt.yticks(np.arange(10, 50, 10), fontweight='bold')
plt.xlabel("Weight", fontweight='bold')
plt.ylabel("MPG", fontweight='bold')
plt.grid(color='grey')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax.set_facecolor("black")
```

```python
X = cars['Weight']
Y = cars['MPG']
s = [X/75]
colors = []
for e in cars["Manufacturer"]:
    if e == "bmw":
        colors.append('#F27831')
    elif e == "ford":
        colors.append('#ADBE0F')
    elif e == "honda":
        colors.append('#0FC37F')
    elif e == "mercedes":
        colors.append('#1CBEF6')  
    else:
        colors.append('#ED8AF8') 
        
figure(figsize=(8, 6), dpi=80)
plt.scatter(X, Y, s=s, alpha=0.5, c=colors)
    
plt.xlim([1500, 5050])
plt.ylim([8, 47])
plt.xticks(np.arange(2000, 6000, 1000))
plt.yticks(np.arange(10, 50, 10))
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.grid(color='white')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax = plt.axes()
ax.set_facecolor("#F6F6F6")
```
